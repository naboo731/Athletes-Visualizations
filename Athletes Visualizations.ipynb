```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import bokeh as bk

%matplotlib notebook
```

```python
data = pd.read_csv('athletes.csv')
data.head()
```

```python
top_medals = data.groupby('nationality')[['gold', 'silver', 'bronze']]\
.sum().sort_values('gold', ascending=False).head(15)

top_medals
```

```python
fig, ax = plt.subplots()
top_medals.plot.bar(color=['gold', 'silver', '#CD7F32'], ax=ax, stacked=True)
```

```python
df = pd.read_csv('athletes.csv')
```

```python
f = df[df['sex'] == 'female']
m = df[df['sex'] == 'male']

tennis_f = f[f['sport'] == 'tennis']
tennis_m = m[m['sport'] == 'tennis']

gymnast_f = f[f['sport'] == 'gymnastics']
gymnast_m = m[m['sport'] == 'gymnastics']
```

```python
fig, ax = plt.subplots()
plots = [(tennis_f, 'red', 'Tennis, women'), 
         (tennis_m, 'blue', 'Tennis, men'), 
         (gymnast_f, 'orange', 'Gymnastics, women'), 
         (gymnast_m, 'green', 'Gymnastics, men')]
for df, color, label in plots:
    df.plot.scatter(ax=ax, x='weight', y='height', alpha=.25, color=color, label=label)
    
ax.set_xlabel('Weight (kg)')
ax.set_ylabel('Height (m)')
ax.set_title('Heights and Weights of Olympic Athletes');
```
